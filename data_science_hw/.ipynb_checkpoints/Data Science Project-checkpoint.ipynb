## Data Types and Missing Values

One quick method to assess the quality of the data is to find the datatypes of each of the columns and the number of missing entries. Already from looking at the dataframe, we can see there might be a problem because the missing values appear to be coded as "Not Available" rather than as `np.nan` (not a number) which means the numeric columns will not be represented as numeric because pandas converts columns with any strings into columns of all strings.

```python
# See the column data types and non-missing values
data.info()
```

Sure enough, there are a number of numeric columns that have been recorded as `object` datatypes. These will have to be converted to `float` datatype before we can do any numerical analysis. We will do that by first converting the "Not Available" entires into `np.nan` which can be interpreted as floats. Then we will convert the columns that contain numeric values (such as square feet or energy usage) into numeric datatypes. This will allow us to then perform visualization and analysis of these columns.

```python
data = data.replace({'Not Available': np.nan})

# Iterate through the columns
for col in list(data.columns):
    # Select columns that should be numeric
    if ('ft²' in col or 'kBtu' in col or 'Metric Tons CO2e' in col or 'kWh' in 
        col or 'therms' in col or 'gal' in col or 'Score' in col):
        # Convert the data type to float
        data[col] = data[col].astype(float)
```

## Missing Values

Now that we have the correct column datatypes, we can lok at the percentage of missing values in each column. Missing values are fine when we do EDA, but they will have to be filled in for machine learning methods. If some columns have an extremely high percentage of missing values, then we might just want to drop those columns because there is not enough information for meaningful conclusions. 

Following is a function that calculates the number of missing values and the percentage of the total values that are missing for each column. As with many tasks in data science, I did not write this function myself, but rather found it online in a [Stack Overflow forum](https://stackoverflow.com/a/39734251)! 

```python
# Function to calculate missing values by column
def missing_values_table(df):
        # Total missing values
        mis_val = df.isnull().sum()
        # Percentage of missing values
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        # Make a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        # Rename the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        # Sort the table by percentage of missing descending
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        # Print some summary information
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        return mis_val_table_ren_columns
```

```python
missing_values_table(data)
```

What columns to retain may be a little arbitrary, but for this project, we will keep any columns that have more than 50% of the values non-missing. In general, be careful about dropping any information because even if it is not there for all the observations, it may still be useful for predicting the target value. In this case, I found that keeping the columns with a high percentage of missing values did not improve model performance and devcided to drop them. 

```python
# Get the missing columns with >50% 
missing_df = missing_values_table(data)
missing_columns = missing_df[missing_df['% of Total Values'] > 50].index
```

```python
# Drop the columns
data = data.drop(columns = list(missing_columns))
```

The rest of the missing values will have to be imputed (filled-in) using an appropriate strategy before doing machine learning. For now, we will move on to exploratory plots in order to find trends, anomalies, or relationships within the data.

# Exploratory Data Analysis

The purpose of Exploratory Data Analysis (EDA) is to find anomalies, patterns, trends, or relationships within the data. These may be interesting by themselves (for example finding a correlation between two variables) or they can be used to inform modeling decisions such as which features to include in a model. In short, the goal of EDA is to determine what our data can tell us! This is an open-ended process that starts out coarse and narrows in as we learn more about the dataset. 

To begin the EDA, we will focus on a single variable, the Energy Star Score, because this is the target for our machine learning models. A simple but effective plot for visualizing the distribution of a single variable is the histogram.

## Univariate Plots

```python
figsize(8, 8)
plt.style.use('fivethirtyeight')
plt.hist(data['ENERGY STAR Score'].dropna(), bins = 100);
plt.xlabel('Score'); plt.ylabel('Count'); 
plt.title('ENERGY Star Score Distribution');
```

Already this is very surprising! The Energy Star Score is a percentile rank that is supposed to give an indicator of the overall energy efficiency of a building. A percentile rank is a relative score meaning that we would expect to see a completely flat distribution with each score making up 1% of the distribution. However, this is completely not the case as we can see that the two most common scores, 0 and 100, make up a disproporationate number of the overall scores. 

Going back to the definition of the Energy Star Score, "1-to-100 percentile ranking for specified building types, calculated in Portfolio Manager, based on self-reported energy usage for the reporting year" we note that this is based on self-reported energy usage. This poses a problem, because a building owner might (unfortunately not everyone is completely honest) be tempted to report lower electricity usage to artificially boost the score of their building. That would explain the high percentage of buildings with a perfect score of 100, but does not explain why so many buildings score at the very bottom! 

To contrast the Energy Star Score, we can look at the Energy Use Intensity (EUI), which is the total energy use divided by the square footage of the building. In this case, the energy usage is not self-reported, so this might be a more objective measure of the energy efficiency of a building. Moreover, this is not a percentile rank, so the absolute values are important and we would expect them to be approximately normally distributed with perhaps a few outliers on the low or high end. 

```python
# Histogram Plot of Site EUI
figsize(8, 8)
plt.hist(data['Site EUI (kBtu/ft²)'].dropna(), 
         range = (0, 400), bins = 20);
plt.xlabel('Site EUI'); 
plt.ylabel('Count'); plt.title('Site EUI Distribution');
```

This plot looks a little less suspicious and is close to normally distributed with a long tail on the right side (it has a positive skew). 

Although this might be a more objective measure, our goal is still to predict the Energy Star Score, so we will move back to examining that variable. Even if the score is not a good measure, it is still our task to predict it, so that is what we will try to do! In the final report back to the company, I will point out this might not be an objective measure, and it might be a good idea to use different metrics to determine the efficiency of a building. If we had more time for this project, it might be interesting to take a look at buildings with scores of 1 and 100 to see if they have anything in common. 

For now, back to the score. In order to look at the effect of categorical variables on the score, we can make a density plot colored by the value of the categorical variable. We can use these plots to see if the categorical variable has any effect on the score. The first plot we will make shows the distribution of scores by the property type. In order to now clutter the plot, we will limit the graph to building types that have more than 80 observations in the dataset. 

```python
# Create a list of buildings with more than 80 measurements
types = data.dropna(subset=['ENERGY STAR Score'])
types = types['Largest Property Use Type'].value_counts()
types = list(types[types.values > 80].index)
```

```python
# Plot of distribution of scores for building categories
figsize(14, 10)

# Plot each building
for b_type in types:
    # Select the building type
    subset = data[data['Largest Property Use Type'] == b_type]
    
    # Density plot of Energy Star scores
    sns.kdeplot(subset['ENERGY STAR Score'].dropna(),
               label = b_type);
    
# label the plot
plt.xlabel('Energy Star Score'); plt.ylabel('Density'); 
plt.title('Density Plot of Energy Star Scores');
```

From this graph, we can see that the building type does have an effect on the score (the negative scores on the graph are an artifact of the [kernel density estimation](https://chemicalstatistician.wordpress.com/2013/06/09/exploratory-data-analysis-kernel-density-estimation-in-r-on-ozone-pollution-data-in-new-york-and-ozonopolis/) procedure).  While I was initially going to focus on the numerical columns for predicting the score, this graph tells us that we should include the property type because this information can be useful for determining the score. As building type is a categorical variable, it will have to be one-hot encoded before we can feed it into a machine learning model. 

We can make the same graph, but this time segmented by borough to see if that has any effect on the Score.

```python
# Create a list of boroughs with more than 80 observations
boroughs = data.dropna(subset=['ENERGY STAR Score'])
boroughs = boroughs['Borough'].value_counts()
boroughs = list(boroughs[boroughs.values > 80].index)
```

```python
# Plot of distribution of scores for boroughs
figsize(14, 10)

# Plot each borough distribution of scores
for borough in boroughs:
    # Select the building type
    subset = data[data['Borough'] == borough]
    
    # Density plot of Energy Star scores
    sns.kdeplot(subset['ENERGY STAR Score'].dropna(),
               label = borough);
    
# label the plot
plt.xlabel('Energy Star Score'); plt.ylabel('Density'); 
plt.title('Density Plot of Energy Star Scores');
```

## Correlations between Features and Target

Next, let's move on to finding correlations between the score and other features. We can calculate the Pearson correlation coefficient, `r` between every numerical variable and the score. The Pearson correlation coefficient is a measure of the strength and direction of a linear relationship between two variables: a value of -1 means the two variables are perfectly negatively linearly correlated and a value of +1 means the two variables are perfectly positively linearly correlated. Although there can be non-linear relationships as well between features and the target, linear relationships are a good way to start exploring trends in the data. 

The following code finds the correlations between the numerical variables and the score (the correlation coefficient can only be calculated with numerical variables). 

```python
# Correlations between the features and score
data.corr()['ENERGY STAR Score'].sort_values()
```

There are no strong lienar correlations, either positive or negative, between the features and the target. Therefore, we might want to look at using transformations of the variables, such as the natural log (`np.log`) or the square root of variables (`np.sqrt`). 

In the following code, we will create log and square root transformations of all the numerical variables and then find correlations. 

```python
# Select numeric columns
numeric_subset = data.select_dtypes('number')

# Create columns with square root and log of numeric columns
for col in numeric_subset.columns:
    # Skip the Energy Star Score column
    if col == 'ENERGY STAR Score':
        next
    else:
        numeric_subset['sqrt_' + col] = np.sqrt(numeric_subset[col])
        numeric_subset['log_' + col] = np.log(numeric_subset[col])
```

```python
# Find correlations with target
print(numeric_subset.corr()['ENERGY STAR Score'].sort_values().head(10))
print('\n', numeric_subset.corr()['ENERGY STAR Score'].sort_values().tail(10))
```

Taking the transformations of the features results in more significant linear correlations. If we want to use a linear model, such as linear regression, we therefore will want to include the log transformations.
