One quick method to assess the quality of the data is to find the datatypes of each of the columns and the number of missing entries. Already from looking at the dataframe, we can see there might be a problem because the missing values appear to be coded as "Not Available" rather than as `np.nan` (not a number) which means the numeric columns will not be represented as numeric because pandas converts columns with any strings into columns of all strings.

Sure enough, there are a number of numeric columns that have been recorded as `object` datatypes. These will have to be converted to `float` datatype before we can do any numerical analysis. ## Missing Values

Now that we have the correct column datatypes, we can lok at the percentage of missing values in each column. Missing values are fine when we do EDA, but they will have to be filled in for machine learning methods. If some columns have an extremely high percentage of missing values, then we might just want to drop those columns because there is not enough information for meaningful conclusions. 

Following is a function that calculates the number of missing values and the percentage of the total values that are missing for each column. As with many tasks in data science, I did not write this function myself, but rather found it online in a [Stack Overflow forum](https://stackoverflow.com/a/39734251)! 

```python
# Function to calculate missing values by column
def missing_values_table(df):
        # Total missing values
        mis_val = df.isnull().sum()
        # Percentage of missing values
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        # Make a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        # Rename the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        # Sort the table by percentage of missing descending
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        # Print some summary information
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        return mis_val_table_ren_columns
```

```python
missing_values_table(data)
```

What columns to retain may be a little arbitrary, but for this project, we will keep any columns that have more than 50% of the values non-missing. In general, be careful about dropping any information because even if it is not there for all the observations, it may still be useful for predicting the target value. In this case, I found that keeping the columns with a high percentage of missing values did not improve model performance and devcided to drop them. 

```python
# Get the missing columns with >50% 
missing_df = missing_values_table(data)
missing_columns = missing_df[missing_df['% of Total Values'] > 50].index
```

# Exploratory Data Analysis

The purpose of Exploratory Data Analysis (EDA) is to find anomalies, patterns, trends, or relationships within the data. These may be interesting by themselves (for example finding a correlation between two variables) or they can be used to inform modeling decisions such as which features to include in a model. In short, the goal of EDA is to determine what our data can tell us! This is an open-ended process that starts out coarse and narrows in as we learn more about the dataset. 

To begin the EDA, we will focus on a single variable, the Energy Star Score, because this is the target for our machine learning models. A simple but effective plot for visualizing the distribution of a single variable is the histogram.

Already this is very surprising! The Energy Star Score is a percentile rank that is supposed to give an indicator of the overall energy efficiency of a building. A percentile rank is a relative score meaning that we would expect to see a completely flat distribution with each score making up 1% of the distribution. However, this is completely not the case as we can see that the two most common scores, 0 and 100, make up a disproporationate number of the overall scores. 

Going back to the definition of the Energy Star Score, "1-to-100 percentile ranking for specified building types, calculated in Portfolio Manager, based on self-reported energy usage for the reporting year" we note that this is based on self-reported energy usage. This poses a problem, because a building owner might (unfortunately not everyone is completely honest) be tempted to report lower electricity usage to artificially boost the score of their building. That would explain the high percentage of buildings with a perfect score of 100, but does not explain why so many buildings score at the very bottom! 

To contrast the Energy Star Score, we can look at the Energy Use Intensity (EUI), which is the total energy use divided by the square footage of the building. In this case, the energy usage is not self-reported, so this might be a more objective measure of the energy efficiency of a building. Moreover, this is not a percentile rank, so the absolute values are important and we would expect them to be approximately normally distributed with perhaps a few outliers on the low or high end.

## Correlations between Features and Target

Next, let's move on to finding correlations between the score and other features. We can calculate the Pearson correlation coefficient, `r` between every numerical variable and the score. The Pearson correlation coefficient is a measure of the strength and direction of a linear relationship between two variables: a value of -1 means the two variables are perfectly negatively linearly correlated and a value of +1 means the two variables are perfectly positively linearly correlated. Although there can be non-linear relationships between the features and targets and correlation coefficients do not account for interactions between features, linear relationships are a good way to start exploring trends in the data that we can then use for selecting the features to employ in our model.

To account for possible non-linear relationships, we can take square root and natural log transformations of the features and then calculate the correlation coefficients with the score. To try and capture any possible relationships between the borough or building type (remember these are categorical variables) and the score we will have to one-hot encode these columns.

In the following code, we take log and square root transformations of the numerical variables, one-hot encode all the categorical variables, limit the data to only buildings with a score, calculate the correlations between all of the features and the score, and display the top 20 most positive and top 20 most negative correlations. This is a lot, but with pandas, it is straightforward to do each step one at a time. 

```python
# Select the numeric columns
numeric_subset = data.select_dtypes('number')

# Create columns with square root and log of numeric columns
for col in numeric_subset.columns:
    # Skip the Energy Star Score column
    if col == 'ENERGY STAR Score':
        next
    else:
        numeric_subset['sqrt_' + col] = np.sqrt(numeric_subset[col])
        numeric_subset['log_' + col] = np.log(numeric_subset[col])

# Display the first five rows
numeric_subset.head()
```

```python
# Select the categorical columns
categorical_subset = data[['Borough', 'Largest Property Use Type']]

# One hot encode
categorical_subset = pd.get_dummies(categorical_subset)

# Display the first five rows
categorical_subset.head()
```

```python
# Join the two dataframes using concat
# Make sure to use axis = 1 to perform a column bind
features = pd.concat([numeric_subset, categorical_subset], axis = 1)

# Drop buildings without an energy star score
features = features.dropna(subset = ['ENERGY STAR Score'])
```

```python
# Find correlations with the score and display
correlations = features.corr()['ENERGY STAR Score'].dropna().sort_values()
```

# Feature Engineering

Guided by the EDA, we will create a set of features that we believe will allow our model to learn how to predict the Energy Star Score. Feature engineering often takes a little domain knowledge as well as using statistical features of the dataset. 

We will do the following feature engineering:

* Select the numerical variables and the two categorical variables identified
* Taking the log transformation of the numerical variables
* One-hot encode the categorical variables
* Remove collinear features
* Split into training and testing sets

The last point deserves a little more explanation which we will cover when we get there! 

```python
# Copy the original data
features = data.copy()

# Select the numeric columns
numeric_subset = data.select_dtypes('number')

# Create columns with log of numeric columns
for col in numeric_subset.columns:
    # Skip the Energy Star Score column
    if col == 'ENERGY STAR Score':
        next
    else:
        numeric_subset['log_' + col] = np.log(numeric_subset[col])
        
# Select the categorical columns
categorical_subset = data[['Borough', 'Largest Property Use Type']]

# One hot encode
categorical_subset = pd.get_dummies(categorical_subset)

# Join the two dataframes using concat
# Make sure to use axis = 1 to perform a column bind
features = pd.concat([numeric_subset, categorical_subset], axis = 1)

features.shape
```

## Remove Collinear Features

Removing collinear features is a method to reduce model complexity by decreasing the number of features and can help to increase model generalization. The objective is to remove features that are highly correlated with one another (not features that are highly correlated with the score!).

As a reminder, following are the most negative correlations with the score.

```python
correlations.head()
```

We notice that the top four all have to do with Energy Use Intensity (EUI) and are slightly different ways of expressing the same measurement. Therefore, these variables are highly collinear. That is, they have a high correlation coefficient between them and we only need to know the value of one in order to encode the information contained in all of the variables. If we remove features that are highly collinear, we are not losing much information because we only need one measure of EUI in order to learn the general relationship between EUI and the score. 

Removing the collinear features allows us to reduce the number of features we use in the model, which should yield a more interpretable model. As long as the features were highly collinear, this should also not hurt performance because by keeping only one of the features, we are keeping the information that is encoded by all of the features. 

As an example of highly collinear variables, let's look at an example of `log_Source EUI` versus `log_Site EUI`. 

```python
figsize(12, 8)
sns.lmplot('log_Source EUI (kBtu/ft²)', 'log_Site EUI (kBtu/ft²)',
          data = features, fit_reg = False);
plt.title('Log Site EUI vs Log Source EUI');
```

We can see that these two features are nearly perfectly positively linearly correlated. If we remove one of these features from our data, we are therefore not removing much information, and this will reduce the complexity of the model.

The following code removes the collinear features based on a threshold we select for the correlation coefficients. It also prints the correlations that it removes so we can see the effect of adjusting the threshold. We will use a threshold of 0.6 which removes one of a pair of features if the correlation coefficient between the features exceeds this value.

```python
def remove_collinear_features(x, corr_val):
    '''
    Obj: Drops features that are strongly correlated to other features.
          This lowers model complexity, and aids in generalizing the model.
    Inputs:
          df: features df (x)
          corr_val: Columns are dropped relative to the corr_val input (e.g. 0.8)
    Output: df that only includes uncorrelated features
    '''
    # Dont want to remove correlations between Energy Star Score
    y = x['ENERGY STAR Score']
    x = x.drop(columns = ['ENERGY STAR Score'])
    
    # Creates Correlation Matrix and Instantiates
    corr_matrix = x.corr()
    iters = range(len(corr_matrix.columns) - 1)
    drop_cols = []

    # Iterates through Correlation Matrix Table to find correlated columns
    for i in iters:
        for j in range(i):
            item = corr_matrix.iloc[j:(j+1), (i+1):(i+2)]
            col = item.columns
            row = item.index
            val = abs(item.values)
            if val >= corr_val:
                # Prints the correlated feature set and the corr val
                print(col.values[0], "|", row.values[0], "|", round(val[0][0], 2))
                drop_cols.append(col.values[0])

    drops = set(drop_cols)
    x = x.drop(columns = drops)
    x = x.drop(columns = ['Site EUI (kBtu/ft²)'])
    x['ENERGY STAR Score'] = y
               
    return x
```

```python
# Remove the collinear features above a specified correlation coefficient
features = remove_collinear_features(features, 0.6)
```

```python
# Remove any columns with all na values
features  = features.dropna(axis=1, how = 'all')
features.shape
```

## Split Into Training and Testing Sets

We will actuallyt split the data into 3 different sets: a set of all building __without__ Energy Star Scores (these will later be used to make predictions for which we do not know the answer), a testing set of 30% of the buildings with scores, and a training set of 70% of the buildings with scores. During training, the objective is to build a model that can learn the mapping from the features to the target on the training dataset. We hope that this mapping then generalizes to new data, which is why we test the model on a testing set where is has not seen the targets. As we know the true answers for the test set, we can then compare the test predictions to the true test targets to gauge how well our model will perform when deployed in the real world. The test error is not a perfect indicator, but if our model has not seen the test set answers, then it will be our best estimate of the error for an application.

Splitting the data into a random training and testing set is simply using scikit-learn. First we extract the buildings with no Energy Star Score and then split the datasets.

```python
# Extract the buildings with no score and the buildings with a score
no_score = features[features['ENERGY STAR Score'].isna()]
score = features[features['ENERGY STAR Score'].notnull()]

print(no_score.shape)
print(score.shape)
```

```python
# Separate out the features and targets
features = score.drop(columns='ENERGY STAR Score')
targets = pd.DataFrame(score['ENERGY STAR Score'])

# Split into 70% training and 30% testing set
X, X_test, y, y_test = train_test_split(features, targets, test_size = 0.3, random_state = 42)

print(X.shape)
print(X_test.shape)
print(y.shape)
print(y_test.shape)
```

We have 2104 buildings with no score, 6749 buildings with a score in the training set, and 2893 buildings with a score in the testing set. Now we can move on to the modeling! 

# Establishing Baseline Metric

Before we get carried away with making machine learning models, we have to figure out a naive baseline to see if machine learning is even applicable to our problem. If the models we build cannot outperform a naive guess (we'll see this implementation in a minute), then we might have to admit that machine learning is not suited for this problem. This could be because we are not using the right models, because we need more data, or because there is a simpler solution that does not require machine learning. Establishing a baseline is crucial so we do not end up building a machine learning model only to realize we can't actually solve the problem.

For a regression task, a good naive baseline is to predict the median value of the target on the training set for all examples on the test set. This is simple to implement and sets a relatively low bar for our models: if they cannot do better than a slightly informed guess, then we will need to rethink our approach. 

## Metric: Mean Absolute Error

There are a number of metrics used in regression, with the most common the root mean-squared error. However, I perform the __mean absolute error__ because it shows how far off our model's estimates are on average from the correct value. According to Andrew Nyg, it is best to use a single real-valued performance metric to assess a model, and so we will use the single mean absolute error as our assessment. 

```python
# Function to calculate mean absolute error
def mae(y_true, y_pred):
    return np.mean(abs(y_true - y_pred))
```

```python
baseline_guess = np.median(y)

print("Baseline Performance on the test set: MAE = %0.4f" % mae(y_test, baseline_guess))
```

This shows our average estimate on the test set is off by over 25 points. The naive method of guessing the median training value provides us with a low baseline for our models to beat!
